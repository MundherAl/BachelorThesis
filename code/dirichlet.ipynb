```python
import pandas as pd
import random as rd
import numpy as np
import matplotlib.pyplot as plt
import ternary

sample_1 = np.random.dirichlet((5,5,5), 10000)
sample_2 = np.random.dirichlet((1,1,1), 10000)
sample_3 = np.random.dirichlet((0.1,0.1,0.1), 10000)

# ternary subplots

figure, ax = plt.subplots(1,3, figsize=(15, 5))
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
tax1 = ternary.TernaryAxesSubplot(ax=ax[0])
tax1.gridlines(multiple=0.1, color='black', alpha=0.5)
tax1.boundary(linewidth=1)
tax1.scatter(sample_1, marker='.', color='#1E88E5', alpha=0.1)
tax1.ticks(axis='lbr', linewidth=1, multiple=0.2, offset=0.02, tick_formats='%.1f')
tax1.clear_matplotlib_ticks()
tax1.set_title(' = (5,5,5)', y=-0.1)
tax2 = ternary.TernaryAxesSubplot(ax=ax[1])
tax2.gridlines(multiple=0.1, color='black', alpha=0.5)
tax2.boundary(linewidth=1)
tax2.scatter(sample_2, marker='.', color='#1E88E5', alpha=0.1)
tax2.ticks(axis='lbr', linewidth=1, multiple=0.2, offset=0.02, tick_formats='%.1f')
tax2.clear_matplotlib_ticks()
tax2.set_title(' = (1,1,1)', y=-0.1)
tax3 = ternary.TernaryAxesSubplot(ax=ax[2])
tax3.gridlines(multiple=0.1, color='black', alpha=0.5)
tax3.boundary(linewidth=1)
tax3.scatter(sample_3, marker='$lol$', color='#1E88E5', alpha=0.1)
tax3.ticks(axis='lbr', linewidth=1, multiple=0.2, offset=0.02, tick_formats='%.1f')
tax3.clear_matplotlib_ticks()
tax3.set_title(' = (0.1,0.1,0.1)', y=-0.1)
plt.show()


```
