```julia
using HTTP
using Gumbo
using JSON

# URL does not change. This is set as a global variable for all functions to use instead of always have to redeclare it.
url = "https://www.scamalert.sg/stories/GetStoryListAjax" 

"""
    Retrieves the content of a page using its number.

    # Arguments
    - 'page::Int': number of page.
"""
function get_page(page)
    r = HTTP.post(url; body=Dict("page" => page, "sortBy" => "Latest"))
    json = JSON.parse(String(r.body))
    json_result = JSON.parse(json["result"])
    return json_result
end

"""
    Retrieves n pages and saves them.
"""
function save_n_pages(delay, total_pages, chunk_size)
    pages = []
    current_page = 1

    for i in 1:(total_pages/chunk_size) # number of chunks
        for j in 1:chunk_size
            push!(pages, get_page(current_page))
            println("Retrieved page number $current_page") # To see progress
            current_page = current_page + 1
            sleep(delay) # wait time between requests
        end
        open("pages.json", "w") do file
            JSON.print(file, pages)
        end
    end

    return pages
end
```

```julia
pages = save_n_pages(90, 590, 59)
```
