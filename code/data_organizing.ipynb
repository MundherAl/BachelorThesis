```python
import pandas as pd
import numpy as np
```

```python
file = open("data/stories.csv", "r")
stories_array = []

for line in file:
    stories_array.append(line)

file.close()
```

```python
file = open("data/scam_alert_story_extraction.log", "r")

# read lines that start with "Error"
error_lines = []
for line in file:
    if line.startswith("Error"):
        error_lines.append(line)

file.close()

error_lines = pd.DataFrame(error_lines, columns=['error'])
error_lines

# regex to extract numbers between "story" and ":"
import re

error_lines['story_id'] = error_lines['error'].apply(lambda x: re.findall(r'story (\d+):', x)[0]).apply(lambda x: int(x) - 1)
error_lines['story_id'] = error_lines['story_id']
error_lines

for missing_story_id in error_lines['story_id']:
    stories_array.insert(missing_story_id, "None")

stories_array_df = pd.DataFrame(stories_array, columns=['story'])
stories_array_df
```
